## 4. Select top 5 numerical features that separate the groups in the "class" column

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import StandardScaler
Data_Cortex_Nuclear = pd.read_csv('Data_Cortex_Nuclear.csv')
Data_Cortex_Nuclear.fillna(0)
```

```python
## We are interested in the top 5 features that best discriminate between the classes
## This is a feature selection problem (as opposed to feature transformation accomplished through PCA)
## Since we didn't have time to cover this during the course, I will give you some hints


# use the mutual_info_classif criterion which measures the mutual information between a numerical feature and a categorical feature (our class column)
# SelectKBest will compute a discrepancy metric (in our case mutual_info_classif) between each numerical column and the class column and return the top k features with the hightest score

numFeatures =5
#featureData =Data_Cortex_Nuclear.loc[:, Data_Cortex_Nuclear.columns!='class'].values  
targetData =Data_Cortex_Nuclear['class']# the class column from the dataframe                                                                                                                         
numericalData=scaler.fit_transform(Data_Cortex_Nuclear.fillna(0).drop(columns=['MouseID', 'Genotype','Treatment','Behavior','class']))                                                                                                                                                                                                    
#print(targetData)
#numericalData = # the standardized numerical data from the previous paragraph
numericalData.shape
best_features = SelectKBest(mutual_info_classif, k=numFeatures).fit_transform(numericalData,targetData)
best_features = pd.DataFrame({'first': best_features[:, 0], 'second': best_features[:, 1], 'third': best_features[:, 2], ' fourth': best_features[:, 3], 'fifth ': best_features[:, 4]})
best_features.shape 
```

```python
print(best_features)
```

## 5. Show the scatter matrix for the selected features (with hue = class column) 

```python

import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale 
numar_comp = 5
pca = PCA(n_components = numar_comp).fit(best_features)
pcaFeatures = pd.DataFrame(pca.transform(best_features))
pcaFeatures['class']=targetData
best_features.shape
```

```python
sns.pairplot(pcaFeatures,hue="class",vars=pcaFeatures.columns[[0,1,2,3,4]])
```

## 6. Perform PCA on all standarized numerical features

```python
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
Data_Cortex_Nuclear= pd.read_csv('Data_Cortex_Nuclear.csv')
Data_Cortex_Nuclear.head(30)
scaler = StandardScaler() 
data_scaled = scaler.fit_transform(Data_Cortex_Nuclear.fillna(0).drop(columns=['MouseID', 'Genotype','Treatment','Behavior','class']))
data_scaled.shape

```

```python
numar_comp = 77
pca = PCA(n_components = numar_comp).fit(data_scaled)
pcaFeatures = pd.DataFrame(pca.transform(data_scaled))
pcaFeatures['class']=targetData
sns.pairplot(pcaFeatures,hue="class",vars=pcaFeatures.columns[[0,3,4]])
```

## 7. Using a bar plot, show the variance explained by the first k principal components (with k = 1 to numberOfFeatures)

```python
figure = plt.figure(figsize = (20,10))
x = np.arange(1,numar_comp+1,1)
height = pca.explained_variance_ratio_.cumsum()
plt.xlabel('Principal component',fontdict = {'fontsize':20}) 
plt.ylabel('Explained variance',fontdict = {'fontsize':20})

plt.bar(x, height)
for i, v in enumerate(height):
    plt.text(x[i] - 0.25, v + 0.01, r'%.2f'%v)
```
